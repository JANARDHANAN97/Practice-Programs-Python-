```python
primeList = []
for i in range(2,1000):
    prime = True
    for j in range(2,int(i/2) +1):
        if(i%j == 0):
            prime = False
            break
    if(prime):
        primeList.append(i)
primeList.sort()
print("Twin Primes are")
for ele in primeList:
    if ((ele-2) in primeList or (ele+2) in primeList):
        print(ele)
            
```
